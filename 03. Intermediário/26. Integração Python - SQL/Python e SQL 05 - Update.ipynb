### Update

```python
import pyodbc

dado_conexao = ('Driver={SQLite3 ODBC Driver};Server=localhost;Database=chinook.db')
conexao = pyodbc.connect(dado_conexao)
print('conexao bem sucedida')

cursor = conexao.cursor()
print('cursor criado')
```

```python
cursor.execute("""
    UPDATE customers
    SET Email = 'luisinh0123g@embraer.com.br'
    WHERE Email = 'luisg@embraer.com.br'
""")
```

```python
cursor.commit()
```

```python
cursor.close()
conexao.close()
```
