```python
from bs4 import BeautifulSoup
import requests

requisicao = requests.get('https://coinmarketcap.com/')
site = BeautifulSoup(requisicao.text, 'html.parser')

print(site.prettify())
```

```python
import re

tbody = site.find('tbody')
linhas = tbody.find_all('tr')

for linha in linhas:
    nome_tag = linha.find(class_="coin-item-name")
    preco_tag = linha.find(string=re.compile(r"\$"))  # r"" evita warning

    if nome_tag and preco_tag:
        print(nome_tag.text.strip())
        print(preco_tag.strip())
        print("-" * 30)
```
