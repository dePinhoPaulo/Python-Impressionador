```python
from bs4 import BeautifulSoup
import requests

requisicao = requests.get('https://coinmarketcap.com/')
site = BeautifulSoup(requisicao.text, 'html.parser')

print(site.prettify())
```

```python
import re

tbody = site.find('tbody')
linhas = tbody.find_all('tr')

for linha in linhas:
    try:
        print(linha.find(class_="coin-item-name").text)
        print(linha.find(string=re.compile(r"\$")))  # r"" evita warning)
    except AttributeError:
        break
```

```python
import re

tbody = site.find('tbody')
linhas = tbody.find_all('tr')

for linha in linhas:
    texto_linha = linha.get_text(separator=';')
    lista_textos = texto_linha.split(';')
    nome = lista_textos[1]
    preco = lista_textos[3]
    print(nome, preco)
```
