## Desafio

- Vamos automatizar a extração de informações de um sistema e envio de um relatório por e-mail
- No nosso caso, para todo mundo conseguir fazer o mesmo programa, o nosso "sistema" vai ser o Gmail, mas o mesmo processo pode ser feito com qualquer programa do seu computador e qualquer sistema
    - Passo 1: Entrar no sistema (entrar no Gmail)
    - Passo 2: Entrar em uma aba específica do sistema onde tem o nosso relatório (Aba Contatos)
    - Passo 3: Exportar o Relatório (Exportar Contatos)
    - Passo 4: Pegar o relatório exportado, tratar e pegar as informações que queremos
    - Passo 5: Preencher/Atualizar informações do sistema (No nosso caso, criar um e-mail e enviar)

```python
import pyautogui
import time

#alerta na tela
pyautogui.alert('Não mexer no teclado nem mouse, automação irá rodar.')

#apertar tecla do windows
pyautogui.press('win')
#pausa 1 seg
time.sleep(1)
#digitar chrome
pyautogui.write('chrome')
#apertar enter
pyautogui.press('enter')

time.sleep(1)

#entrar gmail
pyautogui.write('gmail')
pyautogui.press('enter')

#esperar carregar
time.sleep(1)

#localizar a imagem -> vai te dar 4 informações posicao x, posicao y, largura e altura
pyautogui.useImageNotFoundException(False)
x, y, largura, altura = pyautogui.locateOnScreen('print2.png')
#clicar no meio da imagem
pyautogui.click(x + largura/2, y + altura/2)

#esperar gmail abrirchrome
while not pyautogui.locateOnScreen('logo_gmail.png'):
    time.sleep(2)



#localizar e clicar no menu
x, y, largura, altura = pyautogui.locateOnScreen('print_menu.png')
pyautogui.click(x + largura/2, y + altura/2)

time.sleep(1)

#localizar e clicar nos contatos
x, y, largura, altura = pyautogui.locateOnScreen('print_contatos.png')
pyautogui.click(x + largura/2, y + altura/2)

#esperar contatos abrir
while not pyautogui.locateOnScreen('logo_contatos.png'):
    time.sleep(1)

#localizar e clicar no exportar
x, y, largura, altura = pyautogui.locateOnScreen('exportar.png')
pyautogui.click(x + largura/2, y + altura/2)

time.sleep(1)

#localizar e clicar, confirmar exportação
x, y, largura, altura = pyautogui.locateOnScreen('confirma_exportar.png')
pyautogui.click(x + largura/2, y + altura/2)
```

### Agora vamos escrever o e-mail

```python
import pandas as pd
import pyperclip

df = pd.read_csv(r'C:\Users\Kiko\Downloads\contacts.csv')
df = df.dropna(axis=1)
display(df)

#combinação de teclas
pyautogui.hotkey('ctrl', 'pgup')

for email in df['Phonetic First Name']:
    #localizar e clicar no escrever
    x, y, largura, altura = pyautogui.locateOnScreen('escrever.png')
    pyautogui.click(x + largura/2, y + altura/2)

    #escrever email
    pyautogui.write('gmail')
    #enter
    pyautogui.press('enter')
    #tab para assunto
    pyautogui.press('tab')
    pyautogui.write('Lira Caloteiro')
    #tab para corpo
    texto = """
        Olá, Bom dia!
        Pague as parcelas!!
    """
    pyperclip.copy(texto)
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.hotkey('ctrl', 'enter')
```

### E se eu já tiver com a aba aberta, como que eu coloco ela na frente?

```python
while not pyautogui.locateOnScreen('print'):
    pyautogui.hotkey('alt', 'shift', 'tab')
print('achei o gmail!')
```
